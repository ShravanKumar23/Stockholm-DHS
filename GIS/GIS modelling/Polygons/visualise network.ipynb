```python
import pandas as pd

df = pd.read_excel('Network.xlsx')
df_1 = df[df['Identifier'] == 'Akalla-Brista-Vallsta']
df_1
```

```python
import plotly.graph_objects as go

# Sample data: latitude and longitude coordinates
latitudes = df['lat'] # Example latitudes
longitudes = df['long']  # Example longitudes

# Plotting the points
fig = go.Figure(go.Scattermapbox(
    lat=latitudes,
    lon=longitudes,
    mode='markers',
    marker=dict(size=4, color='blue')
    #line=dict(width=2, color='blue')
))

fig.update_layout(
    mapbox_style="open-street-map",
    mapbox=dict(
        center=dict(lat=59.4298, lon=17.9504),
        zoom=9
    )
)

fig.show()
fig.write_html("map.html")
```
