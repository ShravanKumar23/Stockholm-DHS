```python
import os

# directory/folder path
dir_path = r'OUTPUT_DATA'

# list to store files
res = []

# Iterate directory
for file_path in os.listdir(dir_path):
    # check if current file_path is a file
    if os.path.isfile(os.path.join(dir_path, file_path)):
        # add filename to list
        res.append(file_path)
print(res)
```

```python
import pandas as pd
df = pd.DataFrame()

inputDir = ".\Output_data\\"

for i in res:
    inputPath = os.path.join(inputDir, str(i))
    df1 = pd.read_excel(io=inputPath)
    df = df.append(df1)

df
```

```python
df.to_excel('allgisresults.xlsx')
```
